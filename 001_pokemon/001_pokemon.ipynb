## ライブラリのインポート

```python
import pandas as pd
```

## データ読み込み

```python
df = pd.read_csv('Pokemon.csv')
```

## 詳細情報の確認`.info()`

```python
df.info()
```

## サイズの確認`.shape`

```python
df.shape
```

## 各タイプのポケモンの数

```python
df['Type 1'].value_counts()
```

## タイプ別の平均ステータス

```python
df.groupby('Type 1')[['HP', 'Attack', 'Defense']].mean() \
.sort_values(by='Attack', ascending=False)
```

## 合計ステータス最強TOP10

```python
df['Total'] = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)
df.sort_values(by='Total', ascending=False)[['Name', 'Type 1', 'Total']].head(10)
```

## 伝説のポケモンだけ抽出

```python
legendary = df[df['Legendary'] == True]
legendary[['Name', 'Type 1', 'Attack', 'Total']]
```

## 炎タイプ攻撃TOP5

```python
fire_attack = df[df['Type 1'] == 'Fire'].sort_values('Attack', ascending=False)
fire_attack[['Name', 'Attack']].head(5)
```

## タイプ別ポケモン数

```python
import matplotlib.pyplot as plt

df['Type 1'].value_counts().plot(kind='bar', figsize=(10, 4))
plt.title('Pokemon Type 1')
plt.show()
```

## 攻撃と素早さの相関

```python
plt.scatter(df['Attack'], df['Speed'])
plt.xlabel('Attack')
plt.ylabel('Speed')
plt.title('Attack vs Speed')
plt.show()
```
